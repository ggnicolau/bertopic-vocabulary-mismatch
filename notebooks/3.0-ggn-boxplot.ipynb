```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('ggplot')

# Set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# Pandas dataframe
data = pd.DataFrame({"Baseline": df_baseline['entropy'], "BERTopic": df_result['entropy']})

# Plot the dataframe
ax = data[['Baseline', 'BERTopic']].plot(kind='box', title='IR entropy comparison')

plt.savefig('IR_entropy_comparison.png', bbox_inches='tight', dpi = 180)

# Display the plot
plt.show()

```
