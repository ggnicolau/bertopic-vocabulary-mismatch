```python
import pandas as pd
path = '../models/predictions'
dataset = 'webkb-parsed'
df_baseline = pd.read_parquet(f'{path}/baseline/df_results_baseline_{dataset}.parquet.gzip', engine='pyarrow')
df_result = pd.read_parquet(f'{path}/proposal/df_results_proposal_{dataset}.parquet.gzip', engine='pyarrow')
```

```python
df_baseline.boxplot('entropy')
```

```python
df_result.boxplot('entropy')
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

fig_path = '../reports/figures'

plt.style.use('ggplot')

# Set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# Pandas dataframe
data = pd.DataFrame({"Baseline": df_baseline['entropy'], "BERTopic": df_result['entropy']})

# Plot the dataframe
ax = data[['Baseline', 'BERTopic']].plot(kind='box', title=f'IR entropy comparison {dataset}')

plt.savefig(f'{fig_path}/IR_entropy_comparison_{dataset}.png', bbox_inches='tight', dpi = 180)

# Display the plot
plt.show()
```
