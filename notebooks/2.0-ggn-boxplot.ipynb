```python
import pandas as pd
path = '../models/predictions'
dataset = 'webkb-parsed'
df_baseline = pd.read_parquet(f'{path}/baseline/df_results_baseline_{dataset}.parquet.gzip', engine='pyarrow')
df_mpnet = pd.read_parquet(f'{path}/mpnet/df_results_proposal_{dataset}.parquet.gzip', engine='pyarrow')
df_distiluse = pd.read_parquet(f'{path}/distiluse/df_results_distiluse_{dataset}.parquet.gzip', engine='pyarrow')
```

```python
df_baseline.boxplot('entropy')
```

```python
df_mpnet.boxplot('entropy')
```

```python
df_distiluse.boxplot('entropy')
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

fig_path = '../reports/figures'

plt.style.use('ggplot')

# Set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# Pandas dataframe
data = pd.DataFrame({"baseline": df_baseline['entropy'], "mpnet": df_mpnet['entropy'], "distiluse": df_distiluse['entropy']})

# Plot the dataframe
ax = data[['baseline', 'mpnet', 'distiluse']].plot(kind='box', title=f'IR entropy comparison {dataset}')

plt.savefig(f'{fig_path}/IR_entropy_comparison_{dataset}.png', bbox_inches='tight', dpi = 180)

# Display the plot
plt.show()
```
